```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

data = {
    'Age': ['<=30', '<=30', '31-40', '>40', '>40', '>40'],
    'Income': ['High', 'Medium', 'Medium', 'Low', 'Medium', 'High'],
    'Buy': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)
print(df)
```

```python
le_age = LabelEncoder()
le_income = LabelEncoder()
le_buy = LabelEncoder()

df['Age'] = le_age.fit_transform(df['Age'])
df['Income'] = le_income.fit_transform(df['Income'])
df['Buy'] = le_buy.fit_transform(df["Buy"])


```

```python
x = df[['Age', 'Income']]
y = df[['Buy']]

clf = DecisionTreeClassifier()
clf.fit(x,y)
```

```python
new_data = [[1,0]]
prediction = clf.predict(new_data)
result =  le_buy.inverse_transform(prediction)
print(result)
```

```python
plt.figure(figsize=(8,5))
plot_tree(clf, feature_names=['Age','Income'], class_names=le_buy.classes_, filled=True)
plt.title("Class")
plt.show
```
